```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from keras import backend
```

```python
path=r"C:\Users\Aylin\Desktop\Aylin\makineOgrenmesi\algorithms"
```

```python
data=pd.read_csv(path+"\data.csv")
data
```

```python
label=data.iloc[:,-1:].values
features=data.iloc[:,:-1].values
```

```python
label
```

```python
features
```

```python
X_train,X_test,y_train,y_test=train_test_split(features,label,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
def model_rate(result):
    confmat=confusion_matrix(y_test,result)
    acc=accuracy_score(y_test,result)
    recall=recall_score(y_test,result,average='micro')
    prec=precision_score(y_test,result,average='micro')
    f_measure=f1_score(y_test,result,average='binary')
    recall_sensitivity =recall_score(y_test, result, pos_label=1)
    print('Confusion matrix : ',confmat)
    print('Accuracy classification score : %.3f'%acc)
    print('Recall score : %.3f'%recall)
    print('Precision score : %.3f'%prec)
    print('F-measure score : %.3f'%f_measure)
    print('Sensitivity score : %.3f'%recall_sensitivity)
```

```python
result=dt.predict(X_test)
```

```python
model_rate(result)
```

```python
from keras.models import Sequential
from tensorflow.keras import models, layers
from tensorflow.keras.layers import Dense, Flatten,LSTM,Dropout,Activation
from tensorflow.keras.utils import to_categorical
from keras.backend import clear_session
```

```python
clear_session()
```

```python
input_shape=(X_train.shape[1],1)

model=Sequential()
model.add(LSTM(128,input_shape=input_shape,activation="relu",return_sequences=True))
model.add(Dropout(0.3))

model.add(LSTM(128,activation="relu",return_sequences=False))
model.add(Dropout(0.3))

model.add(Dense(32,activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss='binary_crossentropy',optimizer="adam",metrics=['accuracy'])
```

```python
num_epochs = 100
num_batch_size = 16
history=model.fit(X_train, y_train, batch_size=num_batch_size, epochs=num_epochs, validation_data=(X_test, y_test), verbose=1,shuffle=False)
```

```python
loss, accuracy = model.evaluate(X_train, y_train, verbose=False)
print("Training Accuracy: {:.4f}".format(accuracy))
loss, accuracy = model.evaluate(X_test, y_test, verbose=False)
print("Testing Loss:  {:.4f}".format(loss))
y_pred=model.predict(X_test)
y_tested = np.round(y_test)
y_predicted = np.round(y_pred)
print('Confusion_matrix: ',confusion_matrix(y_tested,y_predicted))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
